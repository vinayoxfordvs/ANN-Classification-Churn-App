```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
## load the  trained model, scaler pickle , one hot encoder
model = load_model('model.h5')

## load the encoder and scaler
with open('onehot_encoder_geo.pkl', 'rb') as file:
    onehot_encodr_geo = pickle.load(file)

with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)

```

```python
##Example input data 
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40, 
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
## One-hot encode Geography
geo_encoded = onehot_encodr_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encodr_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
```

```python
## encode categorical variables

input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
## concatination with one-hot encoded
input_df= pd.concat([input_df.drop("Geography", axis=1),geo_encoded_df], axis=1)
input_df
```

```python
## Scalling the data
input_scaled=scaler.transform(input_df)
input_scaled
```

```python
##Prediction churn


prediction=model.predict(input_scaled)
prediction
```

```python
prediction_probability=prediction[0][0]
prediction_probability
```

```python
if prediction_probability > 0.5:
    print("The customer is likely to churn.")
else:
    print("The customer is unlikely to churn.")
```
